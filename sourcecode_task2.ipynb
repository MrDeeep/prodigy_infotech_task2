```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop(columns="Cabin",axis=1,inplace=True)
```

```python
df
```

```python
df.fillna({'Age': df['Age'].mean()}, inplace=True)
```

```python
df.fillna({'Embarked': df['Embarked'].mode()[0]}, inplace=True)
df.isnull().sum()
```

```python
df
```

```python
df['Survived'].value_counts()
```

```python
sns.countplot(x='Survived', data=df, palette='Set2', hue = 'Survived', legend = True)
plt.title('Count of Survivors vs Non-Survivors')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.show()
```

```python
sns.countplot(x='Survived',hue='Sex',data=df,palette='Set2',legend = True)
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.show()
```

```python
sns.countplot(x='Pclass',hue = 'Pclass',data=df,palette='Set2')
plt.xlabel("Pclass")
plt.ylabel("Number of people")
plt.show()
```

```python
sns.countplot(x='Survived',hue='Pclass',data=df,palette='Set2')
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.show()
```

```python
sns.catplot(x='Pclass', hue='Survived', col='Sex', kind='count', data=df, palette='Set2')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='Embarked',hue='Embarked',data=df,palette='coolwarm',)
plt.xlabel("Embarked")
plt.ylabel("Number of people")
plt.show()
```

```python
sns.countplot(x='Survived',hue='Embarked',data=df,palette='Set2')
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.show()
```

```python
fig,axes = plt.subplots(1, 2, figsize=(12, 6))
sns.countplot(x='SibSp',hue='SibSp',data=df,ax=axes[0],palette='Set2')
sns.countplot(x='Parch',hue='Parch',data=df,ax=axes[1],palette='Set2')
plt.show()
```

```python
sns.countplot(x ='Survived', hue='SibSp',data=df,palette='Set2')
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.title("Survival population of Sibsp")
plt.show()

sns.countplot(x ='Survived',hue='Parch',data=df,palette='Set2')
plt.xticks(ticks=[0,1],labels=['Not survived','survived'])
plt.title("Survival population of Parch")
plt.xlabel("Survival status")
plt.ylabel("Number of people")
plt.show()
```

```python
cut_points = [0, 5, 12, 18, 35, 60, 100]
label_names = ['Infant', 'Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
df['Age_categories'] = pd.cut(df['Age'], bins=cut_points, labels=label_names, right=False)

age_cat_pivot = df.pivot_table(index="Age_categories", values="Survived", aggfunc='mean', observed=False)

colors = ['blue', 'green', 'orange', 'purple', 'pink', 'brown']
fig, ax = plt.subplots()
bars = ax.bar(age_cat_pivot.index.astype(str), age_cat_pivot['Survived'], color=colors)
handles = [plt.Rectangle((0, 0), 1, 1, color=colors[i]) for i in range(len(colors))]
ax.legend(handles, label_names)
ax.set_title('Survival Rates by Age Category')
ax.set_xlabel('Age Category')
ax.set_ylabel('Survival Rate')
plt.xticks(rotation=40)
plt.show()
```
